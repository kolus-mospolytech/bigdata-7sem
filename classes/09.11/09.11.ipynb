```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/moons.csv")
print(df)
print(df.shape)
plt.figure()
plt.scatter(df['x1'], df['x2'], c=df['label'], cmap='rainbow')
plt.axis('equal')
plt.title("Model data")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
import seaborn as sns

plt.figure()
g = sns.jointplot(x=df['x1'],
                  y=df['x2'],
                  kind='scatter',
                  color='gray',
                  edgecolor='black',
                  alpha=0.6, space=0)
g.ax_joint.set_aspect('equal')
plt.title("Model data")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['label'], test_size=0.25, random_state=0)
print(
    X_train.shape, X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn.fit(X_train, y_train)

prediction = knn.predict(X_test)
print(prediction)
print(X_test.assign(predict=prediction))
prediction = knn.predict(X_test)
print(prediction)
print(X_test.assign(predict=prediction))

print(format(knn.score(X_test, y_test)))
```

```python
points_new = pd.DataFrame({'x1': [0, 200, 170], 'x2': [0, 50, 160]})

prediction = knn.predict(points_new)
print(points_new.assign(predict=prediction)) 
```
